```python
# Conversion of JSON to list
GroundTruthList=[]
for json_str in GroundTruth_jsonlist:
    GroundTruthList.append(json.loads(json_str)) 
TrainingDataList=[]
for json_str in TrainingData_jsonlist:
    TrainingDataList.append(json.loads(json_str))
```

```python
# this function counts dissimilarity between two lists of freq dist
def algorithvalue(list1,list2):
    variable1 = 0
    count = 0
    for a in list1:
        for b in list2:
            if a[0] == b[0]:
                variable3 = ((a[1] - b[1])*2/(a[1] + b[1]))**2
                variable1 = variable1 + variable3
                count = count + 1
    finalvalue = variable1 + (len(list1)+len(list2)-count*2)*4
    return finalvalue
```

```python
# to normalize the frequency distribution of word n gram
def NormalizeWordNGram(tokens):
    fd = nltk.FreqDist(tokens).most_common()
    list1 = []
    list2 = []
    list3 = []
    for m in fd:
        list1.append(m[0])
        list2.append(m[1])
    for n in list2:
        n = n/len(tokens)
        list3.append(n)
    return [(list1[i],list3[i]) for i in range(0,len(list1))]
```

```python
# this function will count the dissimilarity for word unigram with profile length L and the data set between x and y
def Word1gram(L,x,y):
    values = []
    for m in range(x,y):
        d = TrainingDataList[m]['pair'][0]
        u = TrainingDataList[m]['pair'][1]
        dtokens = nltk.word_tokenize(d)
        utokens = nltk.word_tokenize(u)
        d1tokens = dtokens[:int(len(dtokens)/2)]
        d2tokens = dtokens[int(len(dtokens)/2):]
        ungram =  NormalizeWordNGram(utokens)[:L]
        d1ngram = NormalizeWordNGram(d1tokens)[:L]
        d2ngram = NormalizeWordNGram(d2tokens)[:L]
        final = (algorithvalue(ungram,d1ngram) + algorithvalue(ungram,d2ngram))/(2*algorithvalue(d1ngram,d2ngram))
        values.append(final)
    return values
```

```python
# this function will count the dissimilarity for word bigram with profile length L and the data set between x and y
def Word2gram(L,x,y):
    values = []
    for m in range(x,y):
        d = TrainingDataList[m]['pair'][0]
        u = TrainingDataList[m]['pair'][1]
        dtokens = list(bigrams(d))
        utokens = list(bigrams(u))
        d1tokens = dtokens[:int(len(dtokens)/2)]
        d2tokens = dtokens[int(len(dtokens)/2):]
        ungram =  NormalizeWordNGram(utokens)[:L]
        d1ngram = NormalizeWordNGram(d1tokens)[:L]
        d2ngram = NormalizeWordNGram(d2tokens)[:L]
        final = (algorithvalue(ungram,d1ngram) + algorithvalue(ungram,d2ngram))/(2*algorithvalue(d1ngram,d2ngram))
        values.append(final)
        return values
```

```python
# this function will count the dissimilarity for word trigram with profile length L and the data set between x and y
def Word3gram(L,x,y):
    values = []
    for m in range(x,y):
        d = TrainingDataList[m]['pair'][0]
        u = TrainingDataList[m]['pair'][1]
        dtokens = list(trigrams(d))
        utokens = list(trigrams(u))
        d1tokens = dtokens[:int(len(dtokens)/2)]
        d2tokens = dtokens[int(len(dtokens)/2):]
        ungram =  NormalizeWordNGram(utokens)[:L]
        d1ngram = NormalizeWordNGram(d1tokens)[:L]
        d2ngram = NormalizeWordNGram(d2tokens)[:L]
        final = (algorithvalue(ungram,d1ngram) + algorithvalue(ungram,d2ngram))/(2*algorithvalue(d1ngram,d2ngram))
        values.append(final)
        return values
```

```python
# this function will give you character n gram of the string and then give you normalised freq dist of that ngram 
# where n is the character length and m is the number of ngrams
def CharacterNgram(text,n,m):
    charngram = [w.lower() for w in [text[i:i+n] for i in range(len(text)-n+1)]]
    fdtextcharngram = nltk.FreqDist(charngram)
    list1 = []
    list2 = []
    list3 = []
    for x in fdtextcharngram.most_common()[:m]:
        list1.append(x[0])
        list2.append(x[1])
    for x in list2:
        x = x/len(text)
        list3.append(x)
    return [(list1[i],list3[i]) for i in range(0,len(list1))]
```

```python
# this function will count the dissimilarity for n character gram with profile length L and the data set between x and y
def Characterngram(n,L,x,y):
    values = []
    for m in range(x,y):
        u = TrainingDataList[m]['pair'][1]
        d1 = TrainingDataList[m]['pair'][0][:int(len(TrainingDataList[m]['pair'][0])/2)]
        d2 = TrainingDataList[m]['pair'][0][int(len(TrainingDataList[m]['pair'][0])/2):]
        ungram = CharacterNgram(u,n,L)
        d1ngram = CharacterNgram(d1,n,L)
        d2ngram = CharacterNgram(d2,n,L)
        final = (algorithvalue(ungram,d1ngram) + algorithvalue(ungram,d2ngram))/(2*algorithvalue(d1ngram,d2ngram))
        values.append(final)
    return values
```
